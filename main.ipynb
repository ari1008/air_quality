```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dataset/air_quality_health_dataset.csv')
df['aqi'].hist(bins=30, edgecolor='black')
plt.title("Distribution de l'Air Quality Index (AQI)")
plt.xlabel("AQI")
plt.ylabel("Fr√©quence")
plt.show()
```

```python
sns.boxplot(x=df['pm2_5'])
plt.title("Boxplot des PM2.5")
plt.show()
```

```python
correlation_matrix = df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Matrice de corr√©lation")
plt.show()
```

```python
sns.scatterplot(data=df, x="temperature", y="aqi")
plt.title("Temp√©rature vs AQI")
plt.show()
```

```python
df["date"] = pd.to_datetime(df["date"])
df_la = df[df["city"] == "Los Angeles"]

plt.plot(df_la["date"], df_la["aqi"])
plt.title("√âvolution de l'AQI √† Los Angeles")
plt.xlabel("Date")
plt.ylabel("AQI")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
